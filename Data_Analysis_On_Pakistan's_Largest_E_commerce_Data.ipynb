```python
import pandas as pd
import matplotlib as plt
%matplotlib inline
import matplotlib.pyplot as pltp
from matplotlib import style
```

```python
df = pd.read_csv("/content/sample_data/Pakistan Largest Ecommerce Dataset.csv")
```

```python
f = pltp.figure()
f.set_figwidth(10)
f.set_figheight(8)
status = df['status'].value_counts().plot(kind = 'line')
pltp.title("Order Status Frequency")
pltp.xlabel("Order Status")
pltp.ylabel("Frequency")
#according to data on large masses buyers are cancelling their orders which shows uncertainity in e- commerce market due to
#insufficient checking tools of product and saler
```

```python
df['payment_method'].value_counts().plot(kind = 'bar')
pltp.title("Payment Method Frequency")
pltp.xlabel("Payment Method")
pltp.ylabel("Frequency")
#according to data on 90% of the customer prefer COD method
```

```python
explode_list = [0,0,1] # teen parts hain jis ko explode karna hai wahan 1 likh do
colors_list = ['green','yellow','red']
df['BI Status'].value_counts().plot(kind = 'pie',
                                    autopct ='%1.0f%%', # Add percentages
                                    figsize = (10,5),#set chart size
                                    shadow = True,  #To give some 3d effect by giving shadow
                                    explode = explode_list,# Explode the part as mentioned in list
                                    colors = colors_list)


```

```python
f = pltp.figure()
f.set_figwidth(10)
f.set_figheight(5)
df['category_name_1'].value_counts().plot(kind = 'bar',color = 'darkblue')
pltp.title("Best selling Categories w.r.t indivisual orders")
pltp.xlabel("Categories")
pltp.ylabel("Frequency")

# According to our analysis most of the individual orders are related to Beauty,Clothing and Fashion
```

```python
df['grand_total']=df['grand_total'].astype('float')
```

```python
gk = df.groupby("Date")["grand_total"].sum()/10**6
style.use('ggplot')
f = pltp.figure()
f.set_figwidth(14)
f.set_figheight(5)
gk.plot(kind = "bar",width =0.2)
pltp.title("Monthly Sales")
pltp.xlabel("Date")
pltp.ylabel("Sales in million")
# According to data in November 2016 most sales have been incurred.
```

```python
df['discount_amount']=df['discount_amount'].astype('float')
dk = df.groupby("category_name_1")["discount_amount"].sum()/10**5
style.use('ggplot')
f = pltp.figure()
f.set_figwidth(10)
f.set_figheight(5)
dk.plot(kind = "bar",color= 'green')
pltp.title("Discount on Categories")
pltp.xlabel("Categories")
pltp.ylabel("Discount amount in million")
# According to data Mpbiles and tablets gives the large discount to customers.
```

```python
# df["qty_ordered"] = df["qty_ordered"].astype('int')
style.use('ggplot')
f = pltp.figure()
f.set_figwidth(10)
f.set_figheight(5)
df.groupby("category_name_1")["qty_ordered"].sum().plot(kind = 'bar',color = 'green')
pltp.title("Best Selling Category w.r.t Quantity ordered")
pltp.xlabel("Categories")
pltp.ylabel("Quantity ordered")
# According to the data the best-selling category is Mobiles and Tables.
```

```python
style.use('ggplot')
f = pltp.figure()
f.set_figwidth(13)
f.set_figheight(8)
status = df.groupby('payment_method')['status'].value_counts().plot(kind = 'bar',color = 'green')

pltp.title("Order Status w.r.t payment methods ")
pltp.xlabel("Order Status")
pltp.ylabel("Frequency")

# This is very important aspect before starting e commerce business to select the transaction method.
#According to our data large no of successful orders are using COD method
# ALERT! According to analysis most of the most cases registered are using payaxis channels which is alarming
```
